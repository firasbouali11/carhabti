# Data Preprocessing

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("cars.csv")
df
```

```python
df.describe()
```

```python
df.reservoir = df.reservoir.apply(lambda x: int(x.replace("L","").strip()) if x != "NC" else np.nan)
```

```python
df.fillna(df.reservoir.mean(),inplace=True)
```

```python
df.reservoir
```

```python
df
```

```python
df["ad_price"] = df["ad_price"].apply(lambda x: float(x.replace("€","").replace(",",".").strip())) 
df.car_model = df.car_model.apply(lambda x: x.split("\n")[0].strip())
```

```python
df
```

```python
dataPair = df[["car_brand","car_model"]]
l = {}
for brand, model in dataPair.values:
    if brand not in l:
        l[brand] = [model]
    if model not in l[brand]: 
        l[brand].append(model)
    
l
```

```python
df.car_color = df.car_color.str.lower()

colors = ["bleu","blanc","noir","pink","rose","gris","rouge","jaune","vert","red","blue","yellow","gold","silver","green","white","black","grey","gray","ceramic","argent","doré"]
def replaceColor(x):
  for color in colors:
    if color in x:
      if color == "red":
        return "rouge"
      elif color == "blue":
        return "bleu"
      elif color == "white":
        return "blanc"
      elif color == "black":
        return "noir"
      elif color == "grey" or color == "gray":
        return "gris"
      elif color == "silver":
        return "argent"
      elif color == "gold":
        return "doré"
      return color
      

df.car_color = df.car_color.apply(lambda x: replaceColor(x))

df.car_color.fillna("noir",inplace=True)
  
```

```python
rawData = df[["car_brand","ad_price","car_color","car_model","reservoir"]]
```

```python
cleaned_data = pd.get_dummies(rawData)
```

```python
price = cleaned_data["ad_price"]
data = cleaned_data.drop("ad_price",1)
```

```python
price
```

```python
data.iloc[:,37:]
```

```python
import pickle

description = data.describe()
l = description.columns.to_list()

brands = [e.replace("car_brand_","") for e in l[1:l.index("car_brand_VOLVO")+1]]
colors = [e.replace("car_color_","") for e in l[l.index("car_color_argent"):l.index("car_color_vert")+1]]
models = [e.replace("car_model_","") for e in l[l.index("car_model_208"):]]

pickle.dump(brands,open("brands.pickle","wb"))
pickle.dump(colors,open("colors.pickle","wb"))
pickle.dump(models,open("models.pickle","wb"))

print(models)
print(brands)
print(colors)
```
